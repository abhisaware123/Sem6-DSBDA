```python
# manual function to calculate the mean, mode, median, minimum, maximum, and standred deviation of the function

def mean_of_dataset(dataset):
    return sum(dataset) / len(dataset)


def median_of_dataset(dataset):
    data = sorted(dataset)  # sorting the dataset
    index = len(dataset) // 2  # checking the middle index
    if len(dataset) % 2 != 0:  # for odd length datset
        return data[index]
    return (data[index - 1] + data[index]) / 2  # for even length datset


def mode_of_dataset(dataset):
    frequency = {}  # created the dictonary which stores the unique value of data and frequency
    for value in dataset: # adding the frequency when to value of data        
        frequency[value] = frequency.get(value, 0) + 1 
    most_frequent = max(frequency.values()) 
    modes = [key for key, value in frequency.items()
             if value == most_frequent] 
    return modes


def standred_deviation(dataset):
    mean = mean_of_dataset(dataset)
    sum = 0
    for i in dataset:
        sum += (i-mean)**2
    std_dev = (sum/(len(dataset)-1))**0.5
    return std_dev


def maximum_in_dataset(dataset):
    max = 0
    for i in dataset:
        if(i > max):
            max = i
    return max


def minimum_in_dataset(dataset):
    max = 1000  # hardcoded this to max can be changed
    for i in dataset:
        if(i < max):
            max = i
    return max
```

```python
# pie-chart to check for species distribution in the dataset in percentile Format
labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
size = data1['Species'].value_counts()
explode = [0, 0, 0] 

plt.figure(figsize=(8,8))
plt.pie(size, explode = explode, labels = labels, shadow = False, autopct = '%.1f%%')
plt.title('Iris Species', fontsize = 25)
plt.legend()
plt.show()
```

```python
arr = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]

# calculating median of every feature when grouped by the species
for i in arr:
    print(f"\nMedian of {i} grouped by Species: ")
    print(data1.groupby(["Species"])[f'{i}'].median())
```

```python
# calculating mean of every feature when grouped by the species
for i in arr:
    print(f"\nMean of {i} grouped by Species: ")
    print(data1.groupby(["Species"])[f'{i}'].mean())
```

```python
# calculating standard deviation of every feature when grouped by the species
for i in arr:
    print(f"\nStandred Deviation of {i} grouped by Species: ")
    print(data1.groupby(["Species"])[f'{i}'].std())
```
